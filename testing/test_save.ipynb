```python
import matplotlib.pyplot as plt
import h5py
import numpy as np
```

```python
def read_hdf5_file(file_path):
    with h5py.File(file_path, 'r') as f:
        oneD = np.array(f['/1D_data'])
        twoD = np.array(f['/2D_data'])
        threeD = np.array(f['/3D_data'])

    return oneD, twoD, threeD

oneD, twoD, threeD = read_hdf5_file("../data/snap0.h5")
```

```python
plt.plot(np.arange(len(oneD)), oneD)
```

```python
plt.imshow(twoD, origin="lower")
```

```python
plt.imshow(np.mean(threeD, axis=2), origin="lower")
```
