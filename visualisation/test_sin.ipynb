```python
import numpy as np

def sinusoidal_sum(x, y, n_modes, m_modes, A, phi):
    """
    Initialize a sum of sinusoidal functions in two directions with different n- and m-modes.

    Parameters:
    x (array-like): x-coordinates.
    y (array-like): y-coordinates.
    n_modes (list of int): List of n-modes to be used in the x-direction.
    m_modes (list of int): List of m-modes to be used in the y-direction.

    Returns:
    numpy.ndarray: Array representing the sum of sinusoidal functions.
    """
    result = np.zeros_like(x, dtype=float)

    for n in n_modes:
        for m in m_modes:
            result += np.sin(2*np.pi*n * x+phi) * np.sin(2*np.pi*m * y)

    result *= A
    return result
```

```python
import matplotlib.pyplot as plt

# Define the grid
x = np.linspace(0, 2*np.pi, 100)
y = np.linspace(0, 2*np.pi, 100)
X, Y = np.meshgrid(x, y)

# Define modes
n_modes = [1]  # Example n-modes
m_modes = [1, 2, 3]     # Example m-modes

# Compute the sum of sinusoidal functions
result = sinusoidal_sum(X, Y, n_modes, m_modes, 1000, phi=np.pi/4)

# Plot the result
plt.imshow(result, extent=(0, 2*np.pi, 0, 2*np.pi), origin='lower')
plt.colorbar()
plt.title('Sum of Sinusoidal Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
